```python
# Plotting the closing prices as a line plot
plt.plot(gme_stock_df['Close'], label='Closing Price', color='purple')  # Plot closing price
plt.title('GME Stock Closing Prices Over Time')  
plt.xlabel('Date') 
plt.ylabel('Closing Price ($)') 
plt.tight_layout()  # Adjust the layout for better fitting
plt.show()  # Display the plot
```

```python
# Make a scatterplot of opening prices vs. closing prices.
plt.scatter(gme_stock_df['Open'], gme_stock_df['Close'], color='red')
plt.title('GME Stock Opening vs. Closing Prices')
plt.xlabel('Opening Price ($)')
plt.ylabel('Closing Price ($)')
plt.show()
```

```python
# Add a new column to your dataframe whose rows represent the change in price during each day and make a histogram of these values.
gme_stock_df['Price Change'] = gme_stock_df['Close'] - gme_stock_df['Open'] # Calculate the price change

#plotting the histogram of the price change
plt.hist(gme_stock_df['Price Change'], bins=20, color='green', edgecolor='black')
plt.title('GME Stock Price Changes')
plt.xlabel('Price Change ($)')
plt.ylabel('Frequency')
plt.show()
```
